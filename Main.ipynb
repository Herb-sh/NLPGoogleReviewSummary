```python
import pandas as pd
import numpy as np
import spacy
from spacy.lang.de.stop_words import STOP_WORDS
```

```python
# Import vom Sprachmodel
nlp = spacy.load('de_core_news_sm')
```

```python
stop_words = STOP_WORDS
```

```python
# Funktionen
def lammatizaion(x):
    
    text = nlp(x)
    
    return [token.lemma_ for token in text]

def delete_stop_words(x):
    
    for word in x:
        if word in stop_words:
            x.remove(word)
    
    return x

def drop_non_words(x):
    
    for word in x:
        if len(word) < 3:
            x.remove(word)
        
    return x
```

# Modell f√ºr Essensbewertung

```python
df = pd.read_csv('/Users/bastian/Documents/HS Aalen/Trimester 6/Text Mining und Web Analytics/Pr√ºfungsleistung/NLPGoogleReviewSummary/data/trainingdata_essen.csv', sep=';')
```

```python
subset = df[['Bewertung', 'Label']]
```

```python
x = list(subset.itertuples(index=False, name=None))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

# Trainingsdaten (Beispiele von Texten, die manuell in die verschiedenen Kategorien eingeteilt wurden)
training_data = x

# Pipeline, die aus einem TfidfVectorizer und einem LinearSVC-Modell besteht
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('svc', LinearSVC())
])

# Trainiere das Modell auf den Trainingsdaten
model.fit([text for text, category in training_data], [category for text, category in training_data])

# Verwende das trainierte Modell, um eine neue Kategorie f√ºr einen gegebenen Text vorherzusagen
def get_category(text):
    predicted_category = model.predict([text])
    return predicted_category[0] if predicted_category else None
```

```python
def get_category(x):
    
    doc = nlp(x)
    for category, keywords_list in keywords.items():
        for keyword in keywords_list:
            if keyword in doc:
                return category
    
    return None
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

# Load the data
data = df

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(df["Bewertung"], df["Label"], test_size=0.2)

# Convert the text to a bag of words representation
vectorizer = CountVectorizer()
train_features = vectorizer.fit_transform(train_data)
test_features = vectorizer.transform(test_data)

# Train a Naive Bayes classifier on the training data
classifier = MultinomialNB()
classifier.fit(train_features, train_labels)

# Test the model on the testing data
accuracy = classifier.score(test_features, test_labels)
print("Accuracy:", accuracy)
```
